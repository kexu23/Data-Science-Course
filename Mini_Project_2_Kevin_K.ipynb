```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("/content/MP2_Data.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(30,30)})
correlation_matrix = df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

Dropping Week1_Stat1 due to zero values and ID because it is just the index

```python
df=df.drop(columns = "Week1_Stat1")
df=df.drop(columns = "ID")
df=df.drop_duplicates()
```

Using elements from week3_MP1 to Week6_Quiz3 with consideration to their correlation to the grade, dropping Week8_Total due to that being the total score of the assignments

```python
df = df.drop(df.iloc[:, 9:45],axis = 1)
```

```python
# Checking columns
df.info()
```

```python
sns.set(rc={'figure.figsize':(10,10)})
correlation_matrix = df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
rows = df.iloc[:, :-1]
cols = df['Grade']
print(rows)
print(cols)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(rows, cols, test_size = 0.2, random_state=3)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
y_train_predict = linear_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("Performance of the model on the training set")
print("\n")
print('RMSE = ', rmse)
print('R2 = ', r2)
```

```python
# model evaluation for testing set
y_test_predict = linear_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("Performance of the model on the testing set")
print("\n")
print('RMSE = ', rmse)
print('R2 = ', r2)
```

```python
# Creating array to use in plot
predictArray = []
for i in y_test_predict:
  x = round(round(i * 2) / 2)
  predictArray.append(x)
print(predictArray)
```

```python
#Creating plot using predictArray
sns.set(rc={"figure.figsize":(10, 10)})
plt.scatter(Y_test, predictArray)
plt.xlabel("Grade")
plt.ylabel("Predicted Grade")
plt.title("Linear Regression Model")
plt.show()
```

```python
# Imports for the ridge model
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

#Imports for validation
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
steps = [
    ('scalar', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('model', Ridge(alpha=3.8, fit_intercept=True))
]
```

```python
ridge_pipeline = Pipeline(steps)
ridge_pipeline.fit(X_train, Y_train)
```

```python
# Predicting R2 Score the Test set results
y_pred_ridge_train = ridge_pipeline.predict(X_train)
r2_score_ridge_train = r2_score(Y_train, y_pred_ridge_train)
```

```python
# Predicting R2 Score the Test set results
y_pred_ridge_test = ridge_pipeline.predict(X_test)
r2_score_ridge_test = r2_score(Y_test, y_pred_ridge_test)
```

```python
# Predicting Cross Validation Score the Test set results
cv_ridge = cross_val_score(estimator = ridge_pipeline, X = X_train, y = Y_train.ravel(), cv = 10)
```

```python
# Predicting RMSE the Test set results
rmse_ridge = (np.sqrt(mean_squared_error(Y_test, y_pred_ridge_test)))
print('R2_score (train): ', r2_score_ridge_train)
print("\n")
print('R2_score (test): ', r2_score_ridge_test)
print("\n")
print("RMSE: ", rmse_ridge)
print("\n")
print('CV: ', cv_ridge.mean())
```

```python
pred = []
for i in y_pred_ridge_test:
  x = round(round(i * 2) / 2)
  pred.append(x)

print(pred)
print(Y_test)
print(y_pred_ridge_test)
```

```python
sns.set(rc={"figure.figsize":(10, 10)})
plt.scatter(Y_test, pred)
plt.xlabel("Grade")
plt.ylabel("Predicted Grade")
plt.title("Ridge Regression Model")
plt.show()
```
